```python
from google.colab import files

data = pd.read_csv('/content/heart.csv')
# Display the first few rows of the dataset
data.head()
```

```python
data.head()
data.info()
data.describe()
```

SEVEN QUESTIONS ON THE HEART DISEASE

```python
#Does age affect the likelihood of heart disease?
# Grouping data by age and calculating the mean heart disease incidence
age_heart_disease = data.groupby('age')['target'].mean()

plt.figure(figsize=(12, 6))
plt.plot(age_heart_disease.index, age_heart_disease.values, marker='o')
plt.title('Heart Disease Incidence by Age')
plt.xlabel('Age')
plt.ylabel('Proportion of Heart Disease')
plt.grid(True)
plt.show()

```

```python
# How does gender influence the occurrence of heart disease?
plt.figure(figsize=(8, 5))
sns.barplot(x='sex', y='target', data=data, hue='sex', palette='coolwarm', legend=False)
plt.title('Heart Disease Incidence by Gender')
plt.xlabel('Gender (1 = Male, 0 = Female)')
plt.ylabel('Proportion of Heart Disease')
plt.show()

```

```python
#Is high cholesterol correlated with heart disease
# Plotting cholesterol levels against heart disease incidence
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='chol', data=data)
plt.title('Cholesterol Levels vs. Heart Disease')
plt.xlabel('Heart Disease (1 = Yes, 0 = No)')
plt.ylabel('Cholesterol (mg/dl)')
plt.show()
```

```python
#What is the relationship between chest pain type and heart disease?
plt.figure(figsize=(8, 5))
sns.barplot(x='cp', y='target', data=data, hue='cp', palette='viridis', legend=False)
plt.title('Heart Disease Incidence by Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Proportion of Heart Disease')
plt.show()
```

```python
 #Does resting blood pressure correlate with heart disease?
 # Plotting resting blood pressure against heart disease incidence
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='trestbps', data=data)
plt.title('Resting Blood Pressure vs. Heart Disease')
plt.xlabel('Heart Disease (1 = Yes, 0 = No)')
plt.ylabel('Resting Blood Pressure (mm Hg)')
plt.show()
```

```python
#is there a relationship between maximum heart rate achieved and heart disease?
# Plotting maximum heart rate against heart disease incidence
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='thalach', data=data)
plt.title('Maximum Heart Rate vs. Heart Disease')
plt.xlabel('Heart Disease (1 = Yes, 0 = No)')
plt.ylabel('Maximum Heart Rate (thalach)')
plt.show()
```

```python
#How does the number of major vessels colored by fluoroscopy relate to heart disease?
# Counting occurrences of heart disease by the number of major vessels
ca_heart_disease = data.groupby('ca')['target'].mean()

plt.figure(figsize=(8, 5))
sns.barplot(x='ca', y='target', data=data, hue='ca', palette='magma', legend=False)
plt.title('Heart Disease Incidence by Number of Major Vessels')
plt.xlabel('Number of Major Vessels (ca)')
plt.ylabel('Proportion of Heart Disease')
plt.show()

```
